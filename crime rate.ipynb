```python
min_val = crime['TOTAL IPC CRIMES'].min()
max_val = crime['TOTAL IPC CRIMES'].max()
range_val = (max_val - min_val) / 4
low = min_val + range_val
medium = low + range_val
high = medium + range_val
```

```python
# Adjusted quantile-based levels for better distribution
q1 = crime['TOTAL IPC CRIMES'].quantile(0.25)
q2 = crime['TOTAL IPC CRIMES'].quantile(0.50)
q3 = crime['TOTAL IPC CRIMES'].quantile(0.75)

def get_crime_level(crime_count):
    if crime_count <= q1:
        return 1  # Low
    elif crime_count <= q2:
        return 2  # Medium Low
    elif crime_count <= q3:
        return 3  # Medium High
    else:
        return 4  # High

crime['CRIME_LEVEL'] = crime['TOTAL IPC CRIMES'].apply(get_crime_level)
```

```python
crime_level_count = crime['CRIME_LEVEL'].value_counts()
crime_level_count
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le1 = LabelEncoder()

# fit and transform the STATE/UT column using the LabelEncoder
crime["STATE/UT"] = le.fit_transform(crime["STATE/UT"])

# fit and transform the DISTRICT column using the LabelEncoder
crime["DISTRICT"] = le1.fit_transform(crime["DISTRICT"])
```

# Logistic Regression

```python
X = crime[['YEAR', 'MURDER', 'ATTEMPT TO MURDER',
       'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER', 'RAPE', 'CUSTODIAL RAPE',
       'OTHER RAPE', 'KIDNAPPING & ABDUCTION',
       'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS',
       'KIDNAPPING AND ABDUCTION OF OTHERS', 'DACOITY',
       'PREPARATION AND ASSEMBLY FOR DACOITY', 'ROBBERY', 'BURGLARY', 'THEFT',
       'AUTO THEFT', 'OTHER THEFT', 'RIOTS', 'CRIMINAL BREACH OF TRUST',
       'CHEATING', 'COUNTERFIETING', 'ARSON', 'HURT/GREVIOUS HURT',
       'DOWRY DEATHS', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY',
       'INSULT TO MODESTY OF WOMEN', 'CRUELTY BY HUSBAND OR HIS RELATIVES',
       'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES',
       'CAUSING DEATH BY NEGLIGENCE', 'OTHER IPC CRIMES', 'TOTAL IPC CRIMES',
       'STATE/UT', 'DISTRICT']]
y = crime['CRIME_LEVEL']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
log_pred = log.predict(X_test)
log_score = log.score(X_test, y_test)
log_prob = log.predict_proba(X_test)[:, 1]
print('Logistic regression score : ',log_score)
print()
print(confusion_matrix(y_test, log_pred))
print()
print(classification_report(y_test, log_pred))
```

```python
cm = confusion_matrix(y_test, log_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC(probability=True)
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_prob = svm.predict_proba(X_test)[:, 1] if svm.probability else None
svm_score = svm.score(X_test, y_test)
print('Support Vector Machine score : ',svm_score)
print()
print(confusion_matrix(y_test, svm_pred))
print()
print(classification_report(y_test, svm_pred))
```
